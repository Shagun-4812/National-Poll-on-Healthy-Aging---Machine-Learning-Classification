# Data Preparation

Preparing various input datasets for experiments:

1. Regular dataset: X
2. Standardised dataset: X_std
3. PCA Reduced dataset: X_pca

```python
# 1) Regular Dataset

X = DocData.drop(['Number of Doctors Visited'], axis=1)  # Stores the input attributes data
y = DocData['Number of Doctors Visited']  # Stores the target variable data

# Importing the Scikit-Learn's train_test_split to split the data into 75% training and 25% testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42) # Stratified splitting

# Printing the shape of the dataset
print(f"\nNumber of samples and attributes of X: {X.shape}\n")
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of the LabelEncoder
le = LabelEncoder()

# Fit the LabelEncoder to the target labels (y) and transform them to numeric labels
y = le.fit_transform(y)

# Print the transformed numeric labels
print(y)
```

```python
print(f'X.shape: {X.shape}')
print(f'y.shape: {y.shape}')
```

```python
print(y)
```

```python
# 2) Standardized Dataset

# Standardizing the data involves subtracting the mean and scaling the values to unit standard deviation
# It is applied to the training dataset to bring values to a standard range.

# Using the StandardScaler class to change the regular data to Standardized data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std = pd.DataFrame(sc.fit_transform(X))

# Using train_test_split to create a new split for the Standardized input data
X_train_std, X_test_std = train_test_split(X_std, test_size=0.25, stratify=y, random_state=42)

# Inspecting the shape of the dataset after Standardization and taking a peak into it
print(f"\nNumber of samples and attributes of X after standardization: {X_std.shape}\n")
X_std.head()
```

```python
# 3) PCA Reduced Dataset

# Using PCA class to reduce dimensionality of the data from 14 to 10
from sklearn.decomposition import PCA
pca_n = PCA(n_components=10)
X_pca = pd.DataFrame(pca_n.fit_transform(X))

X_train_pca, X_test_pca = train_test_split(X_pca, test_size=0.25, stratify=y, random_state=42)

print(f'Number of samples and attributes of X after PCA reduction: {X_pca.shape}')
X_pca.head()
```

# Dimensionality Reduction

### Principal Component Analysis (PCA)

Principal Component Analysis reduces the features of the data by finding their directions of maximum variance and projecting the data along those directions, thus preserving information of the original attributes and also reducing the size of the dataset for better computations and visualisations.

```python
# Here, we reduce the dimensions of the input data from 10 to 3 so that it can be visualised in 3D

from sklearn.decomposition import PCA

pca2 = PCA(n_components=3)
X_2D_pca = pca2.fit_transform(X)

print(f'\nPCA Features: Number of samples and attributes: {X_2D_pca.shape}')
```

3D Point Cloud with respect to target variable - Number of Doctors Visited

```python
# Visualising the 3D point cloud of the PCA reduced dataset against the original target variable 'Number of Doctors Visited'

y_num = DocData['Number of Doctors Visited']

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

ax.scatter(X_2D_pca[:, 0], X_2D_pca[:, 1], y_num, alpha=0.35);

ax.set_xlabel("PCA-1")
ax.set_ylabel("PCA-2")
ax.set_zlabel("Number of Doctors Visited")

plt.title("Visualisation of PCA reduced data in 3D")
plt.tight_layout()
plt.show()
# fig.savefig('PCA visualisation.png')
```

Differentiating data points by colour based on their class label

```python
# Visualising the 3D point cloud of the PCA reduced dataset against the categorical target variable 'Number of Doctors Visited'

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

# Plotting separate 3D scatterplots based on class labels of the target variable 'Number of Doctors Visited'
ax.scatter(X_2D_pca[y == 0, 0], X_2D_pca[y == 0, 1], y_num[y == 0], color="r", alpha=0.45, label='Class 0')
ax.scatter(X_2D_pca[y == 1, 0], X_2D_pca[y == 1, 1], y_num[y == 1], color="b", alpha=0.45, label='Class 1')
ax.scatter(X_2D_pca[y == 2, 0], X_2D_pca[y == 2, 1], y_num[y == 2], color="g", alpha=0.45, label='Class 2')

# Setting axis labels
ax.set_xlabel("PCA-1")
ax.set_ylabel("PCA-2")
ax.set_zlabel("Number of Doctors Visited")

# Adding title and legend
plt.title("Visualisation of PCA reduced data in 3D")
plt.legend()

# Adjusting layout and displaying the plot
plt.tight_layout()
plt.show()

# Saving the figure
# fig.savefig('PCA_visualisation.png')
```

# Baseline Model - Softmax Regression

### Regular Data

```python
# Creating a Softmax regression model and fitting it to the training data

from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression(multi_class='multinomial', solver='lbfgs')
lr_clf.fit(X_train, y_train)

# The model's score on the testing data
print("REGULAR DATA")
lr_score = lr_clf.score(X_test, y_test).round(5)
print('Classification accuracy: ', lr_score)
```

```python
## Confusion Matrix

# Model's predictions on the test data
y_pred = lr_clf.predict(X_test)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

disp_lr = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_true = y_test, y_pred = y_pred))
disp_lr.plot()
plt.title("Confusion Matrix for Softmax Regression with Regular Data")
plt.show()
```

```python
from sklearn.metrics import classification_report

lr_report = classification_report(y_test, lr_clf.predict(X_test))
print(lr_report)
```

```python
# Cross validation of the model using 5 folds in terms of mean +/- std

from sklearn.model_selection import cross_val_score

cv_scores_lr = cross_val_score(lr_clf, X_train, y_train, scoring='f1_weighted', cv=5)
cv_score_lr = (str(cv_scores_lr.mean().round(5)) + ' +/- ' + str(cv_scores_lr.std().round(5)))
print('\n', cv_score_lr)
```

```python
from sklearn.metrics import f1_score

f1_score_lr =  f1_score(y_true=y_test, y_pred=y_pred, average='weighted')
print(f1_score_lr)
```
